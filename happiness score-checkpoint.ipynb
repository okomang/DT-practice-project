```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
y=hap_new.iloc[ : , 0]
print(y)
x=hap_new.iloc[ : , 1:]
print(x)
```

```python
from sklearn.metrics import r2_score
```

```python
def maxr2_score(regr,x,y):
    max_r_score=0
    for r_state in range(42,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        pred=regr.predict(x_test)
        r_score=r2_score(pred,y_test)
        print("R2 score is",r_score,"at",r_state)
        if r_score>max_r_score:
            max_r_score=r_score
            final_r_state=r_state
    print('\n')
    print('Max r_score is',max_r_score,'at',final_r_state)
    return final_r_state
```

```python
lr=LinearRegression()
r_state=maxr2_score(lr,x,y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=60,test_size=0.30)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
from sklearn.linear_model import Lasso
l=Lasso()
parameters={'alpha':[0.001,0.01,0.1,1]}
c=GridSearchCV(l,parameters,cv=10)
c.fit(x_train,y_train)
c.best_params_
```

```python
l=Lasso(alpha=0.01)
r_state=maxr2_score(l,x,y)
```

```python
c=cross_val_score(l,x,y,scoring='r2',cv=5).mean()
print(c)
```

```python
from sklearn import svm
from sklearn.svm import SVR
```

```python
svr=SVR()
parameters={'kernel':['rbf','linear','poly']}
cc=GridSearchCV(svr,parameters,cv=10)
cc.fit(x_train,y_train)
cc.best_params_
```

```python
svr=SVR(kernel='linear')
r_state=maxr2_score(svr,x,y)
```

```python
c=cross_val_score(svr,x,y,cv=5,scoring='r2').mean()
print(c)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
r_state=maxr2_score(dt,x,y)
```

```python
c=cross_val_score(dt,x,y,cv=5,scoring='r2').mean()
print(c)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.neighbors import KNeighborsRegressor
knr=KNeighborsRegressor()
parameters={'n_neighbors':range(0,30)}
q=GridSearchCV(knr,parameters,cv=10)
q.fit(x_train,y_train)
q.best_params_
```

```python
knr=KNeighborsRegressor(n_neighbors=8)
r_state=maxr2_score(knr,x,y)
```

```python
c=cross_val_score(knr,x,y,cv=5,scoring='r2').mean()
print(c)
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
ada=AdaBoostRegressor()
parameters={'learning_rate':[0.001,0.01,0.1],'n_estimators':[10,50,100,150],'base_estimator':[dt,lr,l,svr,knr]}
f=GridSearchCV(ada,parameters,cv=10)
f.fit(x_train,y_train)
f.best_params_
```

```python
ada=AdaBoostRegressor(learning_rate=0.01,base_estimator=l,n_estimators=10)
r_state=maxr2_score(ada,x,y)
```

```python
c=cross_val_score(ada,x,y,cv=5,scoring='r2').mean()
print(c)
```

```python
MAE=[]
MSE=[]
cscore=[]
model=[lr,l,svr,dt,knr,ada]

for m in model:
    m.fit(x_train,y_train)
    predd=m.predict(x_test)
    print(m)
    
    mae=mean_absolute_error(predd,y_test)
    print("MAE is",mae)
    MAE.append(mae)
    
    mse=mean_squared_error(predd,y_test)
    print("MSE is",mse)
    MSE.append(mse)
    
    cv=cross_val_score(m,x,y,cv=5,scoring='r2').mean()
    print('Cross Value Score is',cv)
    cscore.append(cv)
    
    print('\n')
```

```python
FINAL=pd.DataFrame({'Model':['Linear Regression','Lasso','SVR','DecisionTree','K Neighbors','AdaBoost'],
       'R2 Score':['0.9950448308387618','0.9954425098513754','0.9953160933388377','0.8595151061541935','0.9548289272048572','0.995565176661582'],'MAE':MAE,'MSE':MSE,'CROSS VAL SCORE':cscore})
```
